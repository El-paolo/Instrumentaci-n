```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Skip only the second row (index 1) by passing a list to skiprows
df = pd.read_csv('C:/Users/i5-GTX1650S/OneDrive - UNIVERSIDAD ANDRES BELLO/SEMESTRE 9 - Ingeniería Física/Instrumentación/Inf 1/parteb.csv', sep=';', skiprows=[1], nrows=250000, encoding='latin1')# Strip leading and trailing whitespaces from column names
df.columns = df.columns.str.strip()

# Print column names to verify
print(df.columns)
```

```python
# Replace commas with periods in 'CH1' column
df['CH1'] = df['CH1'].astype(str).str.replace(',', '.')
df['CH2'] = df['CH2'].astype(str).str.replace(',', '.')

# Convert 'CH1' to numeric type
df['CH1'] = pd.to_numeric(df['CH1'], errors='coerce')
df['CH2'] = pd.to_numeric(df['CH2'], errors='coerce')

# Now try plotting
ax = df.plot(kind='line', x='X', y='CH1')
df.plot(kind='line', x='X', y='CH2', ax=ax, color='orange', ylabel='Ch1 y Ch2', xlabel='X', title='Ch1 y Ch2 para frec. menor a la frec. angular crítica')

plt.show()
```

```python
# Replace commas with periods in 'CH1' column
df['CH1.1'] = df['CH1.1'].astype(str).str.replace(',', '.')
df['CH2.1'] = df['CH2.1'].astype(str).str.replace(',', '.')

# Convert 'CH1' to numeric type
df['CH1.1'] = pd.to_numeric(df['CH1.1'], errors='coerce')
df['CH2.1'] = pd.to_numeric(df['CH2.1'], errors='coerce')

# Now try plotting
ax = df.plot(kind='line', x='X.1', y='CH1.1')
df.plot(kind='line', x='X', y='CH2.1', ax=ax, color='orange', ylabel='Ch1 y Ch2', xlabel='X', title='Ch1 y Ch2 para frec. similar a la frec. angular crítica')
```
